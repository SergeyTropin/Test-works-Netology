```python
# Для построения гистограммы распределения выразим
# категориальную переменную salary в числовом виде (salary_digit) по правилу:
# low - 0
# medium - 1
# high - 2

def classify_salary(row):

  if row['salary'] in ['low']:
    return int(0)

  elif row['salary'] in ['medium'] :
    return int(1)

  elif row['salary'] in ['high'] :
    return int(2)

  else:
    return None
```

```python
data['salary_digit'] = data.apply(classify_salary, axis=1)
```

```python
# фильтр (не уволились)
df_in_work = data[data['left'] == 0]
```

```python
# 7. Проверить гипотезу, что сотрудники с высоким окладом
# проводят на работе больше времени, чем сотрудники с низким
# окладом
```

```python
# Нулевая гипотеза: Средние двух выборок (время на работе сотрудников с низким и высоким окладом) одинаковые
```

```python
import scipy.stats as st
```

```python
# выборка 1  (не уволились и уровень зарплаты низкий)
low_salary_workers = data[(data['left'] == 0) & (data['salary_digit'] == 0)]
```

```python
x_low = np.array(list(low_salary_workers['average_montly_hours']))
x_low
```

```python
# выборка 2 (не уволились и уровень зарплаты высокий )
high_salary_workers = data[(data['left'] == 0) & (data['salary_digit'] == 2)]
```

```python
x_high = np.array(list(high_salary_workers['average_montly_hours']))
x_high
```

```python
# Функция сравнения дисперсий двух выборок

def check_variance(group1, group2):
    stat, p = st.levene(group1,group2)
    print(f"p = {p:.5f}")
    if p <0.05:
        print("Отклоняем нулевую гипотезу >> Вариация в группах различается")
    else:
        print("Не отклоняем нулевую гипотезу >> Вариация в группах одинаковая")
```

```python
# Сравниваем дисперсии двух выборок, в случае равенства дисперсий
# сможем сравнить средние по критерию Стьюдента

check_variance(x_low, x_high)
```

Вывод из проверки: дисперсии одинаковые

```python
# Критерий Стьюдента (сравнение средних двух выборок)
```

```python
# stat - значение статистики
# p - p-value

stat, p = st.ttest_ind(x_low, x_high)

print(f"Статистика = {stat:.3f}, p = {p:.3f}")

if p > 0.05:
    print('Не отклоняем нулевую гипотезу, средние, вероятно, одинаковые')
else:
    print('Отклоняем нулевую гипотезу, средние, вероятно, различаются')
```

Вывод: Не выявлено статистически значимого отличия в выборках. Вероятно, время, проведенное на работе сотрудниками с низким и высоким окладом, одинаковое

```python
# 8. Рассчитать следующие показатели среди уволившихся и не уволившихся сотрудников (по отдельности):
```

```python
# Доля сотрудников с повышением за последние 5 лет
# Средняя степень удовлетворенности
# Среднее количество проектов
```

```python
def metrics_workers(data, left_ = 0):
    """
    Функция возвращает метрики по сотрудникам
    Параметр left = 0 : сотрудник не уволился,
             left = 1 : сотрудник уволился
    """
    # фильтр  (не уволились)
    if left_ == 0:
      left_str = 'не уволившиеся'
    else:
      left_str = 'уволившиеся'

    df = data[data['left'] == left_]

    raised_df = df[['promotion_last_5years','left']].groupby('promotion_last_5years').count()

    share_raised = raised_df['left'][1]/(raised_df['left'][0] + raised_df['left'][1])
    print(f'Доля сотрудников с повышением среди {left_str} {100*share_raised: .1f} процентов')

    satisf_level = df['satisfaction_level'].mean()
    print(f'Средняя степень удовлетворенности среди {left_str} {satisf_level: .2f}')

    project_mean = df['number_project'].mean()
    print(f'Среднее количество проектов среди {left_str} {project_mean: .2f}')
```

```python
metrics_workers(data, left_ = 0)
```

```python
metrics_workers(data, left_ = 1)
```

```python
# 9.  Разделить данные на тестовую и обучающую выборки
# Построить модель LDA, предсказывающую уволился ли
# сотрудник на основе имеющихся факторов (кроме department и
# salary)
# Оценить качество модели на тестовой выборке
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Целевая переменная y, набор признаков X

X = data[['satisfaction_level', 'last_evaluation', 'number_project', \
       'average_montly_hours', 'time_spend_company', 'Work_accident', \
          'promotion_last_5years']]

y = data['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
lm = LinearRegression()
```

```python
# Обучение модели LDA
lm.fit(X_train, y_train)
```

```python
# Коэффициенты уравнения линейной регрессии, для соответствующих признаков

lm.coef_
```

```python
# Коэффициент смещения уравнения линейной регрессии, равный целевой переменной, когда все признаки равны 0
lm.intercept_
```

```python
# Предсказания модели на тестовой выборке
y_pred = lm.predict(X_test)
y_pred
```

```python
import sklearn.metrics as metrics
```

```python
# Оценка качества модели (метрики)

# коэффициент детерминации
print(f'R2: {metrics.r2_score(y_test, y_pred): .3f}')

# среднеквадратическая ошибка
print(f'MSE: {metrics.mean_squared_error(y_test, y_pred): .3f}')
```

```python
# Модель линейной регрессии с очень низким качеством прогноза, коэффициент детерминации  (R2 = 0.18), что ниже
# минимально необходимого уровня R2 = 0.5
```

```python
# Распределение остатков (величины ошибки в предсказании модели)
sns.histplot((y_test-y_pred), kde=True, bins=20)
```

```python
# Остатки распределены на две группы, каждая из которых имеет близкое к нормальному распределение.
# Предположительно, эти две группы соответствуют двум значениям целевой переменной (0 и 1)
```

```python
# Коэффициенты уравнения линейной регрессии

pd.DataFrame(lm.coef_ , X.columns, columns=['Коэффициенты'])
```
